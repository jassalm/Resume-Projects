```python
import requests      # Библиотека для отправки запросов
import numpy as np   # Библиотека для матриц, векторов и линала
import pandas as pd  # Библиотека для табличек
import time          # Библиотека для тайм-менеджмента
from bs4 import BeautifulSoup
from time import sleep
from tqdm.notebook import tqdm
import requests
import json
from datetime import datetime
from requests_html import HTMLSession
```

```python
result = pd.read_csv('from_api.csv')
```

```python
parsed = pd.DataFrame(columns=['Название', "Организация:", "Описание", "Тип занятости", "Локация", "Ключевые навыки"])

for item in result['alternate_url']:
    try:
        session = HTMLSession()
        r = session.get(item)
        html = r.content
        soup = BeautifulSoup(html, 'html.parser')
        draft = soup.find_all("script", type="application/ld+json" )[0]
        # сделаем из тэга json, чтобы было удобнее обращаться к характеристикам

        json_ld_data = json.loads(draft.string)

        # вытащим характеристики

        title = json_ld_data['title']
        organization = json_ld_data['hiringOrganization']['name']
        description = json_ld_data['description']

        #немного преобразуем время

        employment_type = json_ld_data['employmentType']
        location = json_ld_data['jobLocation']['address']['addressLocality']

        # вытащим текстовое описание без тегов

        json_text = draft.text
        json_data = json.loads(json_text)
        description = json_data.get('description', None)
        description_without_tags = BeautifulSoup(description, 'html.parser').get_text()

        # вытащим ключевые навыки

        d = soup.find_all("span", "bloko-tag__section bloko-tag__section_text")
        skills = [i.string for i in d]

        # заполним датафрейм

        parsed.loc[item] = [title, organization, description_without_tags, employment_type, location, skills]
        print('done')
    except:
        parsed.loc[item] = [None, None, None, None, None, None]
        print('failed')
```
