Выгрузим все по запросу Data Analyst

```python
data = []

for i in range(200):        #выгружаем данные с двухсот страниц
    url = 'https://api.hh.ru/vacancies?text=Data Analyst&area=113&per_page=10&page={}'.format(i)
    response = requests.get(url)
    print(response)
    unhinged_version = response.json()       #вся инфа с i-ой страницы
    data.append(unhinged_version)
```
